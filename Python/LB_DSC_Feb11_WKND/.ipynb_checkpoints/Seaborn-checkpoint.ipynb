```python
import seaborn as sns
```

```python
week_day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,46,49]
sns.barplot(x=week_day,y=temp)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# relation plot
sns.relplot(x='total_bill',y='tip',data=tips,hue='time',style='smoker',size='size',height=10,aspect=2)
```

```python
# relation plot
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='time',style='smoker',size='size')
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.relplot(x=iris.sepal_length,y=iris.petal_width,data=iris,hue='species')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.barplot(x='class',y='survived',data=titanic)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box',hue='size')
```
